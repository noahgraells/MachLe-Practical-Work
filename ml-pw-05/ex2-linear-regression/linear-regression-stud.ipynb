# Ex2 Linear regression optimisation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pandas import ExcelWriter
from pandas import ExcelFile

%matplotlib inline
```

```python
data = pd.read_excel('overfitting.xlsx', sheet_name=None)
data_train = data['train_set']
data_cv = data['cv_set']
X_train = data_train['x']
y_train = data_train['y']
X_test = data_cv['x']
y_test = data_cv['y']
```

```python
def hypothesis(theta,X):
    return np.dot(theta,X)

def cost(X, y, theta):
    N = len(data)
    return 1 / (2 * N) *np.sum(hypothesis(theta, X) - y)**2

def jTheta(hTheta, y):
    N = len(y)
    return 1/2*N * np.sum((hTheta-y)**2)
```

```python
new_x = np.linspace(-1, 11, 100)

cost_train = []
cost_cv = []
for i in range(1, 8):
    plt.figure(0)
    coeff = np.polyfit(X_train, y_train, deg=i)
    
    cost_train.append(jTheta(np.polyval(coeff, X_train), y_train))
    cost_cv.append(jTheta(np.polyval(coeff, X_test), y_test))
    plt.plot(new_x, np.polyval(coeff, new_x))
    plt.scatter(X_train, y_train, color='black')
    plt.figure(1)
```

```python
plt.plot(list(range(1, 8)), cost_train, label='Cost for the training')
plt.plot(list(range(1, 8)), cost_cv, label='Cost for testing')
plt.legend()
```
